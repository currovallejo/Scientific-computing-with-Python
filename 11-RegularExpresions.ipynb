- before we can use we have to import the library using "<font color='green'>import re</font>"
- you can use <font color='green'>re.search()</font> to see if a string matches a regular expression, similar to using the <span style="color: fuchsia;">find()</span> method for strings
- <font color='green'>re.findall()</font> to extract portions of a string that match your regular expression --> combination of <span style="color: fuchsia;">find()</span> and slicing <span style="color: fuchsia;">var[5:10]</span>

**.rstrip()** <br>
the .rstrip() method is used to remove trailing (or rightmost) whitespace characters from a string. Whitespace characters include spaces, tabs, and newline characters. This method does not modify the original string; instead, it returns a new string with the trailing whitespace removed.

### Using <font color='green'>re.search()</font> like <span style="color: fuchsia;">find()</span>

```python
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if line.find('From:') >=0:
        print(line)
```

```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip() 
    if re.search('From:', line):
        print(line)
```

### Using <font color='green'>re.search()</font> like <span style="color: fuchsia;">startswith()</span>

```python
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
```

```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip() 
    if re.search('^From:', line):
        print(line)
```
